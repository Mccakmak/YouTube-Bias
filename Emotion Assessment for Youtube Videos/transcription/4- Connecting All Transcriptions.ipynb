```python
import pandas as pd
```

```python
df_whisper = pd.read_csv("../data/video_transcription_and_translation_whisper.csv")
df_whisper.head(6)
```

```python
df_all = pd.read_csv("../data/video_transcription.csv")
df_all.head(6)
```

```python
# change old transcription which is 'no transcription' to new transcription
for video_id_whisper in df_whisper.video_id:
    new_transcription = df_whisper[df_whisper['video_id'] == 'L0oo8PsUKrA']['transcription'][0]
    df_all.loc[(df_all['video_id'] == video_id_whisper), 'transcript'] = new_transcription
```

```python
# Save
df_all.to_csv("final_video_transcription.csv", index = False)
```
